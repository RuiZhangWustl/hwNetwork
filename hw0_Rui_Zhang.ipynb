```python
def calculate_row_sums_loop(random_2d_array):
    """
    Fill in this function to calculate row sums of 2d_array using a for loop.
    """
    result=[]
    n= len(random_2d_array)
    m=len(random_2d_array[0])
    for i in range(n):
        sumnum=0
        for j in range(m):
            sumnum+=random_2d_array[i][j]
        result.append(sumnum)
    return result
```

### Imports

```python
from operator import itemgetter
import networkx as nx
import wikipedia
```

```python
"""
Construct a network of Wikipedia pages anchored at Complex Networks
"""

def create_wiki_network(seed_term):
    SEED = seed_term.title()

    STOPS = ("International Standard Serial Number",
             "International Standard Book Number",
             "National Diet Library",
             "International Standard Name Identifier",
             "International Standard Book Number (Identifier)",
             "Pubmed Identifier", "Pubmed Central",
             "Digital Object Identifier", "Arxiv",
             "Proc Natl Acad Sci Usa", "Bibcode",
             "Library Of Congress Control Number", "Jstor")

    todo_lst = [(0, SEED)] # The SEED is in the layer 0
    todo_set = set(SEED)   # The SEED itself
    done_set = set()       # Nothing is done yet

    F = nx.DiGraph()
    layer, page = todo_lst[0]

    while layer < 2:
        del todo_lst[0] #(1)
        done_set.add(page)
        print(layer, page) # Show progress

        try: #(2)
            wiki = wikipedia.page(page)
        except:
            layer, page = todo_lst[0]
            print("Could not load", page)
            continue
        
        # perform some filtering based on the page name
        for link in wiki.links: #(3)
            link = link.title()
            if link not in STOPS and not link.startswith("List Of") \
                        and not link.endswith("(Identifier)"):
                if link not in todo_set and link not in done_set: 
                    todo_lst.append((layer + 1, link))
                    todo_set.add(link)
                F.add_edge(page, link)

        layer, page = todo_lst[0] #(4)
    
    # number of nodes and edges before processing
    print("{} nodes, {} edges".format(len(F), nx.number_of_edges(F)))

    # processing (make sure you understand what this does)
    F.remove_edges_from(nx.selfloop_edges(F))
    duplicates = [(node, node + "s") for node in F if node + "s" in F]
    for dup in duplicates:
        F = nx.contracted_nodes(F, *dup, self_loops=False)
    duplicates = [(x, y) for x, y 
                  in [(node, node.replace("-", " ")) for node in F]
                  if x != y and y in F]
    for dup in duplicates:
        F = nx.contracted_nodes(F, *dup, self_loops=False)
    
    # more processing (make sure you understand what this does)
    core = [node for node, deg in dict(F.degree()).items() if deg >= 2]
    G = nx.subgraph(F, core)
    
    # number of nodes and edges after processing
    print("{} nodes, {} edges".format(len(G), nx.number_of_edges(G)))
    
    
#     # We don't need to store this info explicitly.     
#     # set node attributes (here we chosse indegree)
#     indeg_dict = dict(G.in_degree)
#     nx.set_node_attributes(G, indeg_dict, "in_degree")

    return G
    
Wiki_G = create_wiki_network("Complex network")
```
